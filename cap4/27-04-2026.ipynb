```python
#21
from scipy import stats
import numpy as np
#Produtor assume que o seu Produto 

#A vende mais 8% do que um produto B

#de 200 pessoas, 34 preferiram o produto A

#de 150 pessoas, 18 preferiam o produto B

#responder com 0.95 de confiança

nA  = 200; nB =150

obsA = 34; obsB = 18

pA_amostra = obsA/nA
pB_amostra = obsB/nB

confianca= 0.95
alfa = 1- confianca
valor_p = 1 - alfa/2
valor_z = stats.norm.ppf(valor_p,0,1)

varA = (pA_amostra * (1 - pA_amostra)) /nA
varB = (pB_amostra * (1 - pB_amostra)) /nB
dif_variancias = varA + varB
espon = valor_z * np.sqrt(dif_variancias)

lim_inf_IC = (pA_amostra - pB_amostra) - espon
lim_sup_IC = (pA_amostra - pB_amostra) + espon

print(f'{lim_inf_IC}, {lim_sup_IC}')
```

```python
#23

n1 = 100
n2 = 120
conf = 0.90

sigma_1 = 1
sigma_2 = 1.5

media_a_1 = 87.6
media_a_2 = 74.5

var_1 = sigma_1**2 / n1
var_2 = sigma_2**2/n2

desvio_diff = np.sqrt(var_1 + var_2)

valor_p = 1 - ((1- conf) /2)
valor_z = stats.norm.ppf(valor_p,0, 1)

lim_inf_IC = (media_a_1 - media_a_2) - valor_z * desvio_diff
lim_sup_IC = (media_a_1 - media_a_2) + valor_z * desvio_diff


print(f'{lim_inf_IC}, {lim_sup_IC}')
```
