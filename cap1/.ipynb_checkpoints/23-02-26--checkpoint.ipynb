```python
#Exercício 1

valor_emprestimo = float(input('Insira o valor do empréstimo'))
salario = float(input('Insira o seu salário'))
anos = int(input('Número de anos do financiamento'))

prestacaoMensal = valor_emprestimo / (12 * anos)

if prestacaoMensal > salario * 0.3 :
    print(f'O Crédito não foi aprovado')
else:
    print(f'Empréstimo Aprovado')
    print(f'A prestação mensal vai ser de {prestacaoMensal} euros, durante {anos} anos')
```

```python
# Exercício 4
lista = []

valor = float(input("Insira um valor numérico: "))

while valor >= 0:
    
    if valor not in lista:
        lista.append(valor)
        
    valor = float(input("Insira um valor numérico: "))
    
quantidade = len(lista)

if quantidade > 0:
    maior = lista[0]
    menor = lista[0]
    pos_maior = 0
    pos_menor = 0
    
    for i in range(len(lista)):
        if lista[i] > maior:
            maior = lista[i]
            pos_maior = i
        
        if lista[i] < menor:
            menor = lista[i]
            pos_menor = i

    print(f"Número de elementos não repetidos: {quantidade}")
    print(f"Maior número: {maior} Posição: {pos_maior}")
    print(f"Menor número: {menor} Posição: {pos_menor}")
else:
    print("Nenhum número válido foi inserido.")
```

```python
#Exercício 5
lista = []
indice = 0
for i in range(5):
    valor = float(input('Insira um valor númerico: '))
    while indice < len(lista) and valor > lista[indice]:
        indice+=1
    lista.insert(indice, valor)
print("Lista Ordenada: ", lista)
```

```python
#Exercício 7 
import numpy as np
n = int(input('Insira o valor n Matrix: (n x n): '));
matriz = np.zeros((n, n), dtype=int)  # note os parênteses (n, n)


for i in range(n):
    for j in range(n):
        matriz[i][j] = int(input(f'Insira o valor da posição "[{i}][{j}]: '))

print("\nMatriz:")
print(matriz)

soma_pares = np.sum(matriz[matriz % 2 == 0])
print (f'Soma dos valores pares: {soma_pares}')

if n >=3:
    soma_terceira_coluna = np.sum(matriz[: ,2])
    print(f'Soma da terceira coluna: {soma_terceira_coluna}')
else:
    print('A matriz não tem terceira coluna')
if n>=2:
    print(f'O menor valor da segunda linha é: {np.min(matriz[1])}')
else:
    print('A matriz não tem segunda coluna')
```

```python
#Exercício 10
import numpy as np
def maior(variosValoresInteiros):
    print (f'O maior número inteiro é {np.max(variosValoresInteiros)}')

valores = [];

input_inteiro = int(input('Insira um valor inteiro: '))
valores.append(input_inteiro)

while input_inteiro > 0:
    input_inteiro = int(input('Insira um valor inteiro: '))
    valores.append(input_inteiro)

maior(valores)
```
