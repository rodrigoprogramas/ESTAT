```python
#35
# 35.1 CONDIÇÃO FUNDAMENTAL:
# Para que f(x) seja uma função densidade de probabilidade (f.d.p.),
# a área total sob a curva (integral de -infinito a +infinito) 
# deve ser obrigatoriamente igual a 1.

# CÁLCULO POR RAMOS (ÁREAS):
# O domínio da função está dividido em dois intervalos entre 0 e 2.

# Intervalo [0, 1[: f(x) = k
# A área é um retângulo: Base(1) * Altura(k)
# Integral(0 a 1) k dx = k

# Intervalo [1, 2[: f(x) = k(2 - x)
# A área é um triângulo: [Base(1) * Altura(k)] / 2
# Integral(1 a 2) k(2 - x) dx = 0.5k

# RESOLUÇÃO DA EQUAÇÃO:
# Soma das áreas = 1
# k + 0.5k = 1
# 1.5k = 1  => (3/2)k = 1

# RESULTADO FINAL:
# k = 2/3
import scipy.integrate as integrate
print(f"O valor da constante k é: {2/3:.4f} (ou 2/3)")
p_05_1 = integrate.quad(lambda k: 2/3, 0.5,1)
print (f'P(0.5 < X < 1 ) = {p_05_1[0]}')
```

```python
#49
from scipy import stats
#P(X>5) = 1 - P(X<5) 
p1 = 1 - stats.expon.cdf(5,scale =2)
print(f'A probabilidade de não receber chamadas nos próximos 5 segundos é {p1:.3f}')

#X:NÚMERO de chamadas em 5 segundos
#lambda = número médio de chamadas, por segundo
#lambda = 1/beta = 1/2 = 0.5
#x~poisson(5*0.5) = poisson (2.5)
#P(X>=2) = 1 - P(X<=1) = 0.713
beta = 2
llambda = 1 / beta
p2 = 1 - stats.poisson.cdf(1,5*llambda)
print(f'A probabilidade da central receber pelo menos 2 chamadas nos próximos 5 segundos é {p2:.3f}')
```
