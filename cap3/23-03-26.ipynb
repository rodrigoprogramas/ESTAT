```python
#52
#X: Durabilidade do composto quimico , em dias 
#X~Exp(?)
#P(X>50) = 0.10
#0.10 = 1 - P(X<=50) <=> -P(X<=50) = 0.10-1 <=> -P(X<=50) = -0.90
#<=> P(X<=50)= 0.90 <=> F(50) = 0.90
#1-e^(-50/beta) = 0.90 <=> (-50/beta) = ln(0.10) <=> beta = 21.7
#X~Exp(21.7)
#E(X) = 21.7 dias

#P(X>30) = 1- P(X<=30)
from scipy import stats
p_mais_30 = 1 - stats.expon.cdf(30,scale = 21.7)

print(f'A percentagem de duração por mais de 30 dias é {p_mais_30*100:.2f}')

p_menos_10 = stats.expon.cdf(10,scale = 21.7)
print(f'Percentagem de duração menos 10 dias é {p_menos_10}')
#P(10 < X < 30) = F(30) - F(10)
p_entre_10_30 = stats.expon.cdf(30,scale = 21.7) - stats.expon.cdf(10,scale = 21.7)
print(f'Percentagem de duração entre 10 e 30 dias {p_entre_10_30}')
lucro_medio = (p_menos_10 * 5) + (p_entre_10_30 *30) + (p_mais_30 * 100)

print(f'O lucro médio é {lucro_medio}')
```
