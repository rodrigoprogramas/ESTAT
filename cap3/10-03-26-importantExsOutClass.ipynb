```python
#20

#maquina l1 produz 40% e defeito de 20%
#maquina l2 produz 60$ e defeito de 10%
#X: Número de peças defeitosas em caixas de três peças
from scipy import stats
p_tres_pecas_defeituosas_l1 = 0.40 * stats.binom.pmf(3,3,0.2)
p_tres_pecas_defeituosas_l2 = 0.60 * stats.binom.pmf(3,3,0.1)
print(f'A probablidade de haver 3 peças defeitosas na caixa é de {p_tres_pecas_defeituosas_l1 + p_tres_pecas_defeituosas_l2 }')

#P(X>1) = 1 - P(X=0)
p_tres_pecas_defeituosas_l1 = 0.40 * (1 - stats.binom.pmf(0,3,0.2))
p_tres_pecas_defeituosas_l2 = 0.60 * (1 - stats.binom.pmf(0,3,0.1))
print(f'A probablidade de haver pelo menos uma peça defeitosa na caixa é de {p_tres_pecas_defeituosas_l1 + p_tres_pecas_defeituosas_l2 }')

p_peca_defeitosa_e_ser_l1 = (stats.binom.pmf(1,3,0.2) * 0.4) / ((stats.binom.pmf(1,3,0.2) * 0.4) + (stats.binom.pmf(1,3,0.1) * 0.6))

print(f'A probablidade da caixa ºertemcer a linha l1 após verificar que contém uma peça defeituosa é {p_peca_defeitosa_e_ser_l1}')
```
