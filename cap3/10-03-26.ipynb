```python
from scipy import stats

# Probabilidade global de uma peça ser defeituosa
p_defeito = (0.45 * 0.10) + (0.55 * 0.20)  # 0.155

# n = 10 (tamanho da amostra)
n = 10

# Preços
preco_menos_2 = 5.0
preco_2_ou_mais = 2.5

# P(X < 2) = P(X=0) + P(X=1)
# Usando cdf(1) que é a probabilidade acumulada até 1
p_menos_de_duas = stats.binom.cdf(1, n, p_defeito)

# P(X >= 2) = 1 - P(X < 2)
p_duas_ou_mais = 1 - p_menos_de_duas

# Valor Esperado (E = soma de: Preço * Probabilidade)
valor_esperado = (preco_menos_2 * p_menos_de_duas) + (preco_2_ou_mais * p_duas_ou_mais)

print(f"Probabilidade < 2 defeitos: {p_menos_de_duas:.4f}")
print(f"Valor Esperado: {valor_esperado:.2f}€")
```
