```python
#43
import math
#X: Tempo (em unidades de 100 horas) que um estudante demora por ano nas viagens de metro entre casa e escola
#X~Exp(
#P(1 < X < 2) 
p_entre_100_e_200 = integrate.quad(lambda x: (2/9) * x * (3 - x), 1, 2)
print(f'A probabilidade de um estudante demorar entre 100 e 200 horas por ano é {p_entre_100_e_200[0]:.2f}')

#P(X < 2.5) 
p_menos_250 = integrate.quad(lambda x: (2/9) * x * (3-x) ,0,2.5)

print(f'A probabilidade de um estudante demorar menos de 250 horas por ano é {p_menos_250[0]:.2f}')

#P(X < 2.50 | X > 1.00) = P(X<2.50) E (X>1.00) / P(X>1.0) = P(1.00<X<2.5) / P(X>1.00)

p_entre_100_e_250 = integrate.quad(lambda x: (2/9) * x * (3-x) ,1,2.5)

p_mais_100 = integrate.quad(lambda x: (2/9) * x * (3-x) ,1,3) #mais de 100 vai até ao máximo da função

print(f'A probabilidade de um estudante demorar menos de 250 horas por ano sabendo que ja demorou 100 é de {(p_entre_100_e_250[0] / p_mais_100[0]):.2f}')

#43.2 - Derivar para obter a função densidade 0, x<0; 9x^2-2x^3/27 x pertence [0,3], 1 x>3

#E(x) = integral x* f(x)

valor_esperado = integrate.quad(lambda x : x* ((2/9) * x * (3-x)),0,3)


# Valor Esperado do Quadrado E[X^2]
# Função: x^2 * [(2/9) * x * (3-x)]
valor_esperado_quadrado_x2 = integrate.quad(lambda x: (x**2) * (2/9) * x * (3 - x), 0, 3)

# Variância: E[X^2] - (E[X])^2
# Nota: valor_esperado_quadrado_x2[0] extrai o resultado da integral
# valor_esperado[0] extrai a média calculada anteriormente
variancia = valor_esperado_quadrado_x2[0] - (valor_esperado[0]**2)

print(f"Variância: {variancia:.4f}")
print(f'Valor esperado é {valor_esperado[0]} e o desvio padrão é {math.sqrt(variancia)}')

```

```python
#45
from scipy import stats
#X~U(18.5,b)
#media = 20
#20 = 18.5 + b / 2 <=> 40 - 18.5 = b <=> b = 21.5
#X~U(18.5,21.5)
#P(X>21) = 1- p(x<=21)
# Parametros corretos para X ~ U(18.5, 21.5)
a = 18.5
b = 21.5
amplitude = b - a 
media = (a+b) /2
#loc = a, scale = b - a
p_mais_21 = 1 - stats.uniform.cdf(21, loc=a, scale=amplitude)
print(f"Probabilidade: {p_mais_21*100:.1f}")
desvio_padrao = math.sqrt((amplitude**2) / 12)

ponto_corte = media + desvio_padrao
p_excede = 1 - stats.uniform.cdf(ponto_corte, loc=a, scale=amplitude)

print(f"Média: {media} mm")
print(f"Desvio Padrão: {desvio_padrao:.3f} mm")
print(f"Ponto de Corte: {ponto_corte:.3f} mm")
print(f"Percentagem final: {p_excede*100:.1f}%")



p_sucesso = stats.uniform.cdf(20.6, loc=a, scale=amplitude) - stats.uniform.cdf(19.4, loc=a, scale=amplitude)

# 2. Binomial: n=8, k=4, p=0.4
# "Mais de 4" significa 1 - P(X <= 4)
prob_final = 1 - stats.binom.cdf(4, n=8, p=p_sucesso)

print(f"Probabilidade Sucesso (p): {p_sucesso:.1f}") # Deve imprimir 0.4
print(f"Probabilidade Final (>4): {prob_final}%")
```
