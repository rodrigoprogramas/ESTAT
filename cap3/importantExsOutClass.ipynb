```python
#Exercício 7
#7.1
#Acontecimentos
#A: Planta é atacada por fungos
#B: Planta recuperada
#Planta cutsa 1,20, 1,70 se for contaminada e recuperada (50% chance ser recuperada) 
#se nao for recuperada vai para o lixo, venda  planta 3.50
# S = {AB, NA, ANB, NA}
#P(A) = 0.2, P(NA) = 0.8
#P(B|A) = 0.5
#P(A E B ) = P(A)* P(B|A) 
#Lucro médio = soma de cada lucro multiplicado por cada probablidade
preco_planta = 1.20
preco_recuperacao_planta_atacada = 1.70
preco_planta_consumidor = 3.50
lucro_planta_saudavel = preco_planta_consumidor - preco_planta
lucro_planta_fungos_recuperada = preco_planta_consumidor - preco_recuperacao_planta_atacada
lucro_planta_atacada_nao_recuperada =  -(preco_recuperacao_planta_atacada);
p_planta_recuperada = 0.5
p_planta_atacada = 0.2;
p_planta_atacada_recuperada = p_planta_atacada * p_planta_recuperada;
p_planta_atacada_nao_recuperada = p_planta_atacada * p_planta_recuperada; #no caso é a mesma probablidade (50%)
p_planta_saudavel = 0.8;
lucro_medio_por_planta_vendida = (lucro_planta_saudavel*p_planta_saudavel) + (lucro_planta_fungos_recuperada * p_planta_atacada_recuperada) + (lucro_planta_atacada_nao_recuperada * p_planta_atacada_nao_recuperada)
print(f'O lucro médio por cada planta vendida é: {lucro_medio_por_planta_vendida}')

#7.2 10000 * E[Y], calculado no exercício anterior
print(f'O valor esperado para 1000 plantas é {lucro_medio_por_planta_vendida *10000}')
```
