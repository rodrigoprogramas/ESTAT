```python
#54
#X: Tempo de vida de um produto congelado (em condições aceitáveis) em dias
#X ~Exp(60)

#P(X > ? ) = 0.25%

from scipy import stats
media = 60
prob_estragar = 0.25
prazo = stats.expon.ppf(prob_estragar, scale=media)
print(f'O prazo de validade a definir para que mais de 25% das embalagens se estraguem antes do tempo é {prazo:.1f} dias')


#X: Número de embalages fora de prazo retiradas no ultimo dia, em 9 embalagens 
#X~Bi(9,0.25) - Calculado no ex anterior

#P(X> 1/3 das embalagens)
limite_embalagens = (1/3) * 9 

p_mais_1_terco = 1 - stats.binom.cdf(limite_embalagens,9,0.25)

print(f'A probabilidade de mais de 1/3 das embalagens estarem estragad é de {p_mais_1_terco:.2f}')
```

```python
#64

#X: Realização da prova de um atleta, em minutos 
#X~N(10,2^2)

#P(11 < X < 14) = P(X<14) - P(X<11)

media = 12
desvio_padrao = 2
from scipy import stats

p_entre_11_14_minutos = stats.norm.cdf(14,loc=media,scale=desvio_padrao) - stats.norm.cdf(11,loc=media,scale=desvio_padrao)

print(f'A probabilidade do tempo estar entre 11 e 14 minutos é {p_entre_11_14_minutos:.3f}')


#P(10 < X < 14)

p_defirir_media_desvio_p = 1 - (stats.norm.cdf(14,loc=media,scale=desvio_padrao) - stats.norm.cdf(10,loc=media,scale=desvio_padrao))

print(f'A probabilidade de o tempo diferir do valor médio em mais de um desvio padrão é {p_defirir_media_desvio_p:.3f}')
```
