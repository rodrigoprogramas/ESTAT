```python
#48
#X: Duração de certo tipo de brinquedo
#X~U(10,70)

#Durar mais 3/4 do que a duração média 
a = 10
b = 70
amplitude = b - a 
media= a+b/2
#P(X<3/4*40) = P(X>30) = 1 - P(X<=30) 
from scipy import stats
p_brinquedos_dura_mais_media = 1 - stats.uniform.cdf(30,a,amplitude)

print(f'{p_brinquedos_dura_mais_media}')


#P(X>60 | X>30) = P(X>60 E P X >30)/ P(X>30) = P(X>60) / P(X>30) =  1- P(X<=60) / 1- P(X<=30)

p2 = (1- stats.uniform.cdf(60,a,amplitude)) / (1- stats.uniform.cdf(30,a,amplitude))

print(f'{p2}')


#P(|X-45| >= 30) =P(X-45 >= 30 V X-45 <=30) = P(X>=75 V X <=15)


p_sup = 1 - stats.uniform.cdf(75, a, amplitude) 

p_inf = stats.uniform.cdf(15, a, amplitude)

resultado_final = p_sup + p_inf

print(f'P(X >= 75 ou X <= 15) = {resultado_final:.4f}')

```

```python
#51

#X~Exp(2500)

#P(X>3000) = 1- P(X<=3000)

from scipy import stats
media = 2500
p_durar_mais_3000 = 1 - stats.expon.cdf(3000,scale = media)

print(f'A probabilidade de durar mais de 3000 horas é {p_durar_mais_3000:.3f}')

probabilidade_alvo = 0.20

#função inversa, revolver a equação dá nos a duração
duracao_limite = stats.expon.ppf(probabilidade_alvo, scale=media)

print(f'A duração que não é ultrapassada por 20% dos dispositivos é {duracao_limite:.2f} horas.')

n = 12  # total de dispositivos

p_sucesso = stats.expon.sf(3000, scale=media)

prob_mais_metade = stats.binom.sf(6, n, p_sucesso)
prob_mais_metade = 1- stats.binom.cdf(6,n,p_sucesso)
print(f"Probabilidade de mais de 6 funcionarem: {prob_mais_metade:.4f}")
```

```python
# 69
from scipy import stats
import numpy as np


# X: Quantidade de areia transportada numa pá
# X ~ N(U, 1)
# 16% = (P > 4)

variancia_uni = 1
desvio_padrao = np.sqrt(variancia_uni) 
prob_maior_4 = 0.16 

# 1. Encontrar a média real (u) através da informação de que 16% > 4kg
# Usamos 'ppf' para achar o Z-score correspondente a 84% de probabilidade acumulada
z_score = stats.norm.ppf(1 - prob_maior_4) 

# Se Z = (X - u) / sigma, então u = X - (Z * sigma)
media_calculada = 4 - (z_score * desvio_padrao)

print(f"Média calculada (u): {media_calculada:.2f} kg")


probabilidade_menor_3 = stats.norm.cdf(3, loc=media_calculada, scale=desvio_padrao)

print(f"Probabilidade de transportar menos de 3kg: {probabilidade_menor_3 * 100:.1f}%")

n = 20

media_total = n * media_calculada
variancia_total = n * variancia_uni
sigma_total = np.sqrt(variancia_total)

probabilidade_69_2 = 1 - stats.norm.cdf(62, loc=media_total, scale=sigma_total)

print(f"Probabilidade de terem ficado mais de 62kg: {probabilidade_69_2:.4f}")
```
