```python
#Exercício 13
#13.1
#x = 0, n=6, p =0.04
from scipy import stats

p1 = stats.binom.pmf(0,6,0.04)
print(f'A percentagem de lotes que são vendidos sem inspeção de todos os elementos são {p1:.3f}')


#13.2
#P(X>=1) = 1 - P(X = 0)
p2 = 1 - stats.binom.pmf(0,6,0.04)
print(f'A probabilidade de ser necessário uma inspeção total do lote {p2:.3f}')
```

```python
#Exercícios 15
#15.1
#X  quantidade de bovinos contaminados em 5
#P(X= 5 | X>=1) = P(X=5 E x>= 1) / P(X>=1) = P(X=5) / 1- P(X=0)
from scipy import stats
q_bovinos_contaminados_ser_5 = stats.binom.pmf(5,5,0.2)
q_bovinos_maior_igual_um = 1 - stats.binom.pmf(0,5,0.2)
p1 = q_bovinos_contaminados_ser_5 / q_bovinos_maior_igual_um
print(f'A probabilidade de todos estarem contaminados, sabendo-se que pelo menos um deles está é {p1}')
#15.2
#P(X >= 3) = 1 - P(X<=2) = 1- (P(X=2) + P(X=1) + P(X=0))
p_bovinos_2 = stats.binom.pmf(2,5,0.2)
p_bovinos_1 = stats.binom.pmf(1,5,0.2)
p_bovinos_0 = stats.binom.pmf(0,5,0.2)
p_bovinhos_mais_metade = 1 - (p_bovinos_2 + p_bovinos_1 + p_bovinos_0)
print(f'A probabilidade de mais de metade estar contaminado é {p_bovinhos_mais_metade}')
```

```python
#Exercício 23
#23.1
#X: Número de particulas emitidas por milissegundo lambda = 2
#P(X=2)
from scipy import stats
lambd_a = 2
particulas = 2

d_poisson = stats.poisson(lambd_a)
p_particulas_desejadas= d_poisson.pmf(particulas)
ppd = stats.poisson.pmf(particulas, lambd_a)

print(f'A probabilidade de 2 particulas serem emitidas em 2 ms é {p_particulas_desejadas} ou {ppd:.3f}')

#23.2
#P(X=4) 2*(lambda=2) para que a taxa seja 4 por milessegundo
lambd_a = (lambd_a) * 2
particulas = 4

d_poisson = stats.poisson(lambd_a)
p_particulas_desejadas= d_poisson.pmf(particulas)
ppd = stats.poisson.pmf(particulas, lambd_a)

print(f'A probabilidade de 4 particulas serem emitidas em 4 ms é {p_particulas_desejadas} ou {ppd:.3f}')
#23.3
#P(X>=3) = 1- P(X<=2) = 1- ( P(X=2) + P(X=1) + P(X=0) )
lambd_a = 4
p_2_particulas = stats.poisson.pmf(2, lambd_a)
p_1_particulas = stats.poisson.pmf(1,lambd_a)
p_0_particulas = stats.poisson.pmf(0,lambd_a)
p_3_ou_mais_particulas = 1 - ( p_2_particulas + p_1_particulas + p_0_particulas)

print(f'A probabilidade de serem emitidas pelo menos três particulas é {p_3_ou_mais_particulas:.3f}')
```

```python
#Exercício 25
#lambda = 2
#X: Número de carros que chegam ao parque de estacionamento em cada dia
#25.1 P(X>3) = 1 - P(X<=3) = 1 - ( P(X=3)+ P(X=2) + P(X=1) +P(X=0)
from scipy import stats
p_3_veiculo = stats.poisson.pmf(3,2)
p_2_veiculo = stats.poisson.pmf(2,2)
p_1_veiculo = stats.poisson.pmf(1,2)
p_0_veiculos = stats.poisson.pmf(0,2)
p_enviar_veiculos = 1 - (p_3_veiculo + p_2_veiculo + p_1_veiculo + p_0_veiculos)

print(f'A probabilidade num dado dia se enviar veiculos para outro parque é {p_enviar_veiculos:4f}')

#25.2
#P(X <= K) >= 0.98 - usar a funç\ao inversa do pmf 
capacidade_necessaria = stats.poisson.ppf(0.98,2)

print(f'Para permitir recolher todos os veiculos que cheguem em 98% dos dias, as instalações devem ser aumentadas em {capacidade_necessaria}') 
```
