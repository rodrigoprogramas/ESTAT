```python
#4
#X:Tempo de fabrico de determinadas peças em minutos
#X~U(5,15)
#p = P(X>10) = 1- p(X<=10)
from scipy import stats
import numpy as np

a, b = 5, 15
mu = (a + b) / 2
var_pop = (b - a)**2 / 12

n1 = 100
n2 = 50
delta_t = 15 / 60  #15 segundos para 0.25 minutos

mu_d = 0  # 10 - 10
var_d = (var_pop / n1) + (var_pop / n2)
std_d = np.sqrt(var_d)

# Queremos P(D > 0.25)
prob = stats.norm.sf(delta_t, loc=mu_d, scale=std_d)
print(f'O tempo médio de cada peça é {(5+15)/2}')
print(f" A probabilidade do tempo médio de fabrico de 100 dessas peças exceder o tempo médio de fabrico de 50 dessas peças, em pelo menos 15 segundos é {prob:.4f}")
```

```python
#11
#75% dos residentes da periferia de acordo
#60% dos residentes do centro da cidade de acordo
n1= 50
n2 = 100
#P1-P2~N(O.75-0.60, 0.75(1-0.75)/50+ 0.60(1-0.60/100)
#P(P1-P2<0.10) 
import math
from scipy.stats import norm

p1 = 0.75
n1 = 50
p2 = 0.60
n2 = 100


mu = p1 - p2

# 2. Desvio padrão da diferença (sigma)
# Fórmula: sqrt( (p1*(1-p1)/n1) + (p2*(1-p2)/n2) )
variancia = (p1 * (1 - p1) / n1) + (p2 * (1 - p2) / n2)
sigma = math.sqrt(variancia)

# 3. Valor limite que queremos testar (10%)
limite = 0.10


z = (limite - mu) / sigma

# 5. Cálculo da probabilidade P(Z <= z)
probabilidade = norm.cdf(z)

# Output formatado
print(f"Média: {mu:.4f}")
print(f"Desvio Padrão : {sigma:.6f}")
print(f"Z: {z:.6f}")
print(f"A probabilidade é: {probabilidade:.4f}")
```
